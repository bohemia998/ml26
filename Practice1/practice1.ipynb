```python
import pandas as pd
```

```python
college = pd.read_csv('College.csv')
```

```python
college.head()
```

### Top 3 colleges by enrollement

```python
college.sort_values(by='Enroll', ascending=False).head(3)
```

### Top 3 Public vs Private College Enrollement

```python
college.loc[college['Private'] == 'Yes'].sort_values(by='Enroll', ascending=False).head(3)[['Name', 'Enroll']]
```

```python
college.loc[college['Private'] == 'No'].sort_values(by='Enroll', ascending=False).head(3)[['Name', 'Enroll']]
```

###  Colleges Sorted by Enrollement Rate 

```python
college['enrollment_rate'] = college['Enroll'] / college['Accept'] * 100
college.sort_values(by='enrollment_rate', ascending=False).head(5)[['Name', 'enrollment_rate']]
```
